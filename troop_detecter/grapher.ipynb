```python
# method to read a text file
def read_file(file_path):
    with open(file_path, "r") as file:
        return file.read()
```

```python
def plot_box_on_image(annotation_file_path, image_file_path):
    annotation_line = read_file(annotation_file_path).replace("\n", "")

    parts = annotation_line.split(" ")

    label = parts[0]

    bbox = [float(parts[1]), float(parts[2]), float(parts[3]), float(parts[4])]

    width = 420
    height = 633

    var1 = bbox[0] * width  # left
    var2 = bbox[1] * height  # y
    width = bbox[2]*width
    height = bbox[3]*height

    x_adjust = width/2
    y_adjuist = height/2

    bbox = [var1+x_adjust - width, var2 + y_adjuist- height, var1+x_adjust, var2+y_adjuist]
    print(bbox)

    print(f"[{annotation_line}]")

    from PIL import Image, ImageDraw
    from IPython.display import display

    # method to draw a bounding box on an image
    def draw_bbox(image_path, bbox, color="red", thickness=5):
        image = Image.open(image_path)
        draw = ImageDraw.Draw(image)
        draw.rectangle(bbox, outline=color, width=thickness)
        return image

    image_with_bbox = draw_bbox(image_file_path, bbox)
    display(image_with_bbox)


for i, annotation_path in enumerate(annotation_paths):
    image_path = image_paths[i]
    plot_box_on_image(annotation_path, image_path)
```
